# 방문자 및 거주민 대비 수용 가능 인원수 시각화하기(관광객 반영 수용률 확인)

* 제주시, 서귀포시, 전체(제주도) 기준으로 시각화

```python
df = pd.read_csv('그래프용.csv', encoding='cp949')
```

```python
df
```

```python
# 그래프 그리기
ax = df[["방문자 대비 수용 불가능 수", "거주민 대비 수용가능 수"]].plot.barh(color=['skyblue', 'lightcoral'], figsize=(10, 6))

plt.title('지역별 수용 불가능 및 거주민 대비 수용가능 수')
plt.xlabel('수용 수')
plt.ylabel('지역')

# y축 레이블을 지역 이름으로 설정
ax.set_yticklabels(df['지역'])

plt.show()
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6))  # 그래프 크기 설정

# 방문자 대비 수용가능 수 그래프 그리기
plt.subplot(2, 1, 1)  # 2x1 그리드 중 첫 번째 서브플롯
plt.bar(df['지역'], df['방문자 대비 수용 불가능 수'], color='skyblue')
plt.title('방문자 대비 수용가능 수')
plt.xlabel('지역')
plt.ylabel('수용가능 수')

# 거주민 대비 수용가능 수 그래프 그리기
plt.subplot(2, 1, 2)  # 2x1 그리드 중 두 번째 서브플롯
plt.bar(df['지역'], df['거주민 대비 수용가능 수'], color='lightcoral')
plt.title('거주민 대비 수용가능 수')
plt.xlabel('지역')
plt.ylabel('수용가능 수')

plt.tight_layout()  # 서브플롯 간 간격 조정
plt.show()
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6))  # 그래프 크기 설정

# 지역별 방문자 대비 수용가능 수와 거주민 대비 수용가능 수를 하나의 그래프에 표시
plt.bar(df['지역'], df['방문자 대비 수용 불가능 수'], color='skyblue', label='방문자 대비 수용가능 수')
plt.bar(df['지역'], df['거주민 대비 수용가능 수'], color='lightcoral', label='거주민 대비 수용가능 수', alpha=0.7)  # alpha는 투명도 조절

plt.title('지역별 수용가능 수')
plt.xlabel('지역')
plt.ylabel('수용가능 수')
plt.legend()  # 범례 표시

plt.show()
```
