```python
df = pd.read_csv('제주특별자치도_주제3_상권분석을 위한 유동인구에 따른 카드 매출액 변화 데이터 활용_매쉬업결과_20201215.csv', encoding = 'cp949')
df.head()
```

```python
df.info()
```

```python
splited = df['년월'].str.split('-', expand=True)
splited.columns = ["년", "월"]
# splited=splited.drop(['none'],axis=1) #"none" 열 삭제
df=pd.concat([df, splited], axis=1) #df 데이터에 splited 데이터
df.head()
```

```python
df2019 = df[df['년'] == '2019']
df2019.head()
```

```python
df2019.info()
df2019.shape
```

```python
print(df2019['읍면동명'].unique())
print(df2019['시도명'].unique())
```

```python
print(df2019[df2019['시도명'] == '서귀포시']['읍면동명'].unique())
print(df2019[df2019['시도명'] == '제주시']['읍면동명'].unique())
```

```python
df2019['거주인구'] = df2019['거주인구'].astype(int)
df2019['근무인구'] = df2019['근무인구'].astype(int)
df2019['방문인구'] = df2019['방문인구'].astype(int)
```

```python
df2019.info()
df2019.head()
```

```python
nowork = df2019.drop(['업종명', '데이터기준일자', '년월', '년'], axis=1)
nowork = nowork.reset_index(drop = True)
nowork.head()
```

```python
grouped_df = nowork.groupby(['월', '시도명', '읍면동명', '성별']).agg({
    '이용자수': 'sum',
    '이용금액': 'sum',
    '거주인구': 'sum',
    '근무인구': 'sum',
    '방문인구': 'sum'
}).reset_index()
grouped_df
```

```python
# pivot_table을 사용하여 월별 읍면동별 성별 통계 계산
# 월별 평균 데이터 계산
avg_df = df.groupby('읍면동명').agg({
    '이용자수': 'mean',
    '이용금액': 'mean',
    '거주인구': 'mean',
    '근무인구': 'mean',
    '방문인구': 'mean'
}).reset_index()

# 결과 출력
avg_df.info()
```

```python
avg_df['거주인구'] = avg_df['거주인구'].astype(int)
avg_df['근무인구'] = avg_df['근무인구'].astype(int)
avg_df['방문인구'] = avg_df['방문인구'].astype(int)
avg_df['이용금액'] = avg_df['이용금액'].astype(int)
avg_df['이용자수'] = avg_df['이용자수'].astype(int)
avg_df
```

```python
avg_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 히트맵 생성
plt.figure(figsize=(8, 6))
sns.set(font_scale=1)  # 히트맵의 글꼴 크기 조절
sns.heatmap(avg_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, square=True)

plt.title('상관 히트맵')
plt.show()
```

```python
# 읍면동명별 이용자수 막대 그래프
plt.figure(figsize=(12, 6))

plt.bar(avg_df['읍면동명'], avg_df['이용자수'])
plt.xlabel('읍면동명')
plt.ylabel('이용자수')
plt.title('읍면동명별 이용자수')
plt.xticks(rotation=45)
plt.show()
```

```python
# 산점도 그래프 그리기
plt.figure(figsize=(10, 6))
plt.scatter(avg_df['이용자수'], avg_df['이용금액'], c='b', marker='o', alpha=0.7)  # c는 색상, marker는 마커 모양, alpha는 투명도 조절
plt.xlabel('이용자수')
plt.ylabel('이용금액')
plt.title('읍면동별 이용자수와 이용금액의 산점도')
plt.grid(True)  # 그리드 표시
plt.show()
```
